```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import iisignature  # For computing the signature

# Generate synthetic correlated asset data
T = 500  # Number of time steps
t = np.linspace(0, 2*np.pi, T)


# Generate X1 as a Geometric Brownian Motion (GBM)
def generate_gbm(T, mu=0.0005, sigma=0.01, S0=100):
    """Generate a Geometric Brownian Motion (GBM) path."""
    dt = 1  # Assume daily time steps
    W = np.random.randn(T) * np.sqrt(dt)  # Brownian motion increments
    X1 = S0 * np.exp(np.cumsum((mu - 0.5 * sigma**2) * dt + sigma * W))  # GBM formula
    return X1

# Set parameters
mu = 0.0005  # Drift term (mean return)
sigma = 0.01  # Volatility
S0 = 100  # Initial price

# Generate GBM for X1
X1 = generate_gbm(T, mu, sigma, S0)

correlation_noise = 1

X2 = correlation_noise * np.random.randn(T) + X1  # Asset 2 (Correlated with Asset 1)

# Compute Pearson Correlation
pearson_corr, _ = pearsonr(X1, X2)

# Prepare data for signature computation (reshape as (time, features))
path = np.vstack((X1, X2)).T  # Shape (T, 2)


s = iisignature.prepare(2,2)
logsignature = iisignature.logsig(path,s)
levy_area = logsignature[-1]  # Last element corresponds to Levy area

# Print results
print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Second-Level Signature (Levy Area): {levy_area:.3f}")

# Plot the generated paths
plt.figure(figsize=(10, 4))
plt.plot(X1, X2, label="(X1, X2) - Correlated Assets")
plt.xlabel("Asset 1")
plt.ylabel("Asset 2")
plt.title(f"Correlated Asset Paths\nPearson: {pearson_corr:.3f}, Levy: {levy_area:.3f}")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Drop NaN values from rolling correlation (first few values will be NaN due to window size)

def adf_test(ts):

    # Perform Augmented Dickey-Fuller (ADF) test for stationarity
    adf_result = adfuller(ts)

    # Print test results
    print("Augmented Dickey-Fuller Test Results:")
    print(f"ADF Statistic: {adf_result[0]:.4f}")
    print(f"p-value: {adf_result[1]:.4f}")
    print("Critical Values:")
    for key, value in adf_result[4].items():
        print(f"   {key}: {value:.4f}")

    # Interpretation
    if adf_result[1] < 0.05:
        print("\nThe rolling correlation is likely stationary (p < 0.05).")
    else:
        print("\nThe rolling correlation is likely non-stationary (p >= 0.05).")

```

```python
# Define function to compute the second-level log-signature (Levy area)
def compute_rolling_logsig_levy_area(X1, X2, window_size):
    """Compute rolling log-signature level 2 (Levy area) using a moving window."""
    levy_areas = []
    
    # Prepare logsignature calculation
    s = iisignature.prepare(2, 2)
    
    for i in range(len(X1) - window_size + 1):
        # Extract windowed path
        path = np.vstack((X1[i:i+window_size], X2[i:i+window_size])).T
        
        # Compute logsignature
        logsig = iisignature.logsig(path, s)
        
        # Extract Levy area (last component of logsignature level 2)
        levy_areas.append(logsig[-1])
    
    # Pad with NaNs to align with time series length
    levy_areas = [np.nan] * (window_size - 1) + levy_areas
    return np.array(levy_areas)

for window_size in range(5,200,20):


    # Compute rolling log-signature level 2 (Levy area)
    rolling_levy_area = pd.Series(compute_rolling_logsig_levy_area(X1, X2, window_size=window_size))

    # Plot rolling log-signature Levy area
    plt.figure(figsize=(10, 4))
    plt.plot(rolling_levy_area, label=f"Rolling Levy Area; window={window_size}", color='purple')
    plt.xlabel("Time")
    plt.ylabel("Levy Area")
    plt.title("Rolling Log-Signature Level 2 (Levy Area) for X1 (GBM) and X2")
    plt.legend()
    plt.show()
    
    adf_test(rolling_levy_area.dropna())
```
